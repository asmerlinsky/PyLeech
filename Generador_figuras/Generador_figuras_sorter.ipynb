```python
import PyLeech.Utils.burstStorerLoader as burstStorerLoader
import PyLeech.Utils.burstUtils as burstUtils
burst_object = burstStorerLoader.BurstStorerLoader(filename="2018_12_13_0015.pklspikes", mode='load')
```

```python
fig, ax = burstUtils.plotFreq(burst_object.spike_freq_dict, color_dict=None, template_dict=burst_object.template_dict,
                                  scatter_plot=True, outlier_thres=3.5, draw_list=[3, 8, 13, 15, 18, 21], ms=4, optional_trace=[time[::5], NS[::5]], sharex=ax1)
for a in fig.get_axes():
    try:
        a.get_legend().remove()
    except AttributeError:
        pass
# fig.legend()
fig.get_axes()[-1].set_xlabel("Time (s)")
fig.get_axes()[3].set_ylabel("Frequency (Hz)")
```

```python
fig.savefig("freq_plot.png", dpi=600, transparent=True)
```

```python
import PyLeech.Utils.NLDUtils as NLD
import numpy as np
import scipy.signal as spsig
binning_dt = 0.1
spike_kernel_sigma = 1

spike_freq_array = burstUtils.processSpikeFreqDict(burst_object.spike_freq_dict, binning_dt,
                                                   selected_neurons=[3, 8, 13, 15, 18, 21],
                                                   time_length=burst_object.time[-1])

kernel = NLD.generateGaussianKernel(sigma=spike_kernel_sigma, time_range=20, dt_step=binning_dt)
smoothed_sfd = {}
for key, items in spike_freq_array.items():
    smoothed_sfd[key] = np.array([items[0], spsig.fftconvolve(items[1], kernel, mode='same')])
    
    
kernel = NLD.generateGaussianKernel(sigma=spike_kernel_sigma, time_range=15, dt_step=1/fs)
conv_NS = spsig.fftconvolve(NS, kernel, mode='same')
```

```python
fig1, ax1 = burstUtils.plotFreq(smoothed_sfd, color_dict=None, template_dict=burst_object.template_dict,
                                  scatter_plot=False, outlier_thres=None, draw_list=[3, 8, 13, 15, 18, 21], ms=4, optional_trace=[time[::5], conv_NS[::5]], sharex=ax)
for a in fig1.get_axes():
    try:
        a.get_legend().remove()
    except AttributeError:
        pass

fig1.get_axes()[-1].set_xlabel("Time (s)")
fig1.get_axes()[3].set_ylabel("Frequency (Hz)")
```
