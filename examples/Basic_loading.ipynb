## Basic loading of .abf files

```python
import os
from neo.io import AxonIO
print(AxonIO.mode)
```

```python
print(os.getcwd())
```

```python
#loading file
reader1 = AxonIO("test1.abf")
reader2 = AxonIO("test2.abf")
#reading blocks
blocks1 = reader1.read()
blocks2 = reader2.read()
#getting segments (can be retrieved from reader or from blocks)
seg1 = blocks1[0].segments[0]
seg2 = blocks2[0].segments[0]
```

```python
print("\tblock info:\n\nall attrs:")
for item in blocks1[0]._all_attrs: print(item)
print("\nchild containers:")
print(blocks1[0]._child_containers)

print("\n\n\tsegment info:\n\nall attrs:")
for item in blocks1[0].segments[0]._all_attrs: print(item)
print("\nchild containers:")
for item in blocks1[0].segments[0]._child_containers: print(item)

print("\n\n\tchannel indexes info:\n\nall attrs:")
for item in blocks2[0].channel_indexes[0]._all_attrs: print(item)
print("\nchild containers:")
for item in blocks2[0].channel_indexes[0]._child_containers: print(item)
    
print("\n\n\tanalog signal info:\n\nall attrs:")
for item in blocks1[0].segments[0].analogsignals[0]._all_attrs: print(item)

```
